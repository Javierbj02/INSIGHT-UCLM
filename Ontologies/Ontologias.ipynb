```python
from owlready2 import World, get_ontology, sync_reasoner

soma_url = "https://ease-crc.github.io/soma/owl/current/SOMA.owl"

soma = get_ontology(soma_url).load()
```

```python
num_classes = len(list(soma.classes()))
num_properties = len(list(soma.object_properties()))
num_individuals = len(list(soma.individuals()))

print(num_classes, num_properties, num_individuals)

print("\n")

for c in list(soma.classes())[:10]:
    print(c)

print("\n")

for c in list(soma.object_properties())[:10]:
    print(c)

print("\n")

for c in list(soma.individuals())[:10]:
    print(c)
```

```python
for c in list(soma.classes()):
    print(c)
```

```python
collision = soma.search(label="*Collision*")

if collision:
    print("Collision classes found:")
    for collision in collision:
        print(collision)
else:
    print("No Collision class found.")
```

```python
with soma:
  sync_reasoner()
```

```python
collision_iri = "http://www.ease-crc.org/ont/SOMA.owl#Collision"
collision_class = soma.search_one(iri=collision_iri)
collision_class
```

```python
def clase_jerarquia(clase):
    print("Clase:", clase)
    if clase.is_a:
        print("Hija de:", clase.is_a[0])
        print("\n")
        clase_jerarquia(clase.is_a[0])

clase_jerarquia(collision_class)
```

```python
def clases_raiz(ontology):
  root_classes = []
  subclass_counts = {}

  all_classes = list(ontology.classes())
  all_superclasses = set()

  for cls in all_classes:
      all_superclasses.update(cls.is_a)

  for cls in all_classes:
      if cls not in all_superclasses:
          root_classes.append(cls)

  return root_classes

root_classes = clases_raiz(soma)

print("Clases raiz:")
for root_class in root_classes:
    print(root_class)
```
