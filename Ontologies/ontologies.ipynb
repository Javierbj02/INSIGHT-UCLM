Simple queries

```python
"""Simple queries can be performed with the .search() method of the ontology. It expects one or several keyword arguments. The supported keywords are:
iri, for searching entities by its full IRI
type, for searching Individuals of a given Class
subclass_of, for searching subclasses of a given Class
is_a, for searching both Individuals and subclasses of a given Class
subproperty_of, for searching subproperty of a given Property
any object, data or annotation property name

Special arguments are:
_use_str_as_loc_str: whether to treats plain Python strings as strings in any language (default is True)
_case_sensitive: whether to take lower/upper case into consideration (default is True)
_bm25: if True, returns a list of (entity, relevance) pairs instead of just the entities (default is False)
The value associated to each keyword can be a single value or a list of several values. A star * can be used as a jocker in string values."""
```

```python
print(onto.search(iri = "*")) # returns all entities in the ontology
print(onto.search(iri = "*owl")) # returns all entities in the ontology with IRI ending with "owl"
```

# SOMA

```python
from owlready2 import *

onto = get_ontology("https://ease-crc.github.io/soma/owl/current/SOMA.owl").load()

print("Ontología cargada:", onto)
```

```python
# Listar todas las clases

for cls in onto.classes():
    print(cls) # Tenemos SOMA y DUL. Esto es porque la ontología SOMA hereda de DUL. DUL es la ontología base de SOMA.
```

```python
# Listar propiedades
for prop in onto.properties():
    print(prop) # Todas las propiedades de la ontología SOMA. Las propiedades son las relaciones entre las clases. Por ejemplo, "hasPart" es una propiedad que relaciona dos clases.
```

```python
# Explorar individuos
for ind in onto.individuals():
    print(ind) # Todos los individuos de la ontología SOMA. Los individuos son las instancias de las clases. Por ejemplo, "Soma" es un individuo de la clase "SOMA".
```

```python
# Para saber todos los individuos y sus clases, se puede hacer lo siguiente:
for ind in onto.individuals():
    print(ind, ind.is_a) # Todos los individuos y sus clases. Por ejemplo, "Soma" es un individuo de la clase "SOMA".
```

```python
# Realizar consultas
collision_class = onto.search_one(iri="*Collision") # Busca un individuo con el IRI "Collision". Esto es útil para buscar individuos específicos en la ontología.
print(collision_class) # Imprime el individuo encontrado. Si no se encuentra, devuelve None.

if collision_class:
    collisions = list(collision_class.instances())
    print("Instancias de la clase Collision:")
    for instance in collisions:
        print(instance)
        
else:
    print("No se encontró la clase Collision.")
```

```python
# Realizar consultas
soma_class = onto.search_one(iri="*ExecutionStateRegion")
print(soma_class,"\n") 


classes = list(soma_class.instances())
print("Instancias de la clase Collision:")
for instance in classes:
    print(instance)
        
```

```python
physical_systems = list(onto.ExecutionStateRegion.instances())
print("Instancias de ExecutionStateRegion:")
for system in physical_systems:
    print(system)
```

```python
# Ejemplo: buscar individuos con un valor específico en alguna propiedad. rdfs:label es una propiedad que contiene el nombre de la clase o individuo.
resultados = onto.search(label="affects")
for item in resultados:
    print(item)
```

```python
# Crear una nueva clase que herede de una existente (por ejemplo, PhysicalSystem)

class NuevoSistema(onto.ExecutionStateRegion):
    pass

# Crear una instancia de la nueva clase
sistema_instancia = NuevoSistema("mi_nuevo_sistema") # 
print("Nueva instancia creada:", sistema_instancia)

# Guardar la ontología modificada en un archivo
onto.save(file="SOMA_modificada.owl", format="rdfxml")

# ver nueva ontología - NuevoSistema
onto = get_ontology("SOMA_modificada.owl").load()
for cls in onto.classes():
    print(cls) # Todas las clases de la ontología SOMA. La nueva clase "NuevoSistema" debería aparecer aquí.
    
# ver nueva instancia - mi_nuevo_sistema
for ind in onto.individuals():
    print(ind)
```

```python
print("Antes de ejecutar el razonador:")
mi_nuevo_sistema = onto.search_one(iri="*mi_nuevo_sistema")
print("Clases de mi_instancia:", mi_nuevo_sistema.is_a)
```

```python
#Podar la ontología para eliminar las clases que no se utilizan. Esto es útil para reducir el tamaño de la ontología y mejorar su rendimiento.

onto = get_ontology("SOMA_modificada.owl").load()

clases_sin_instancias = [cls for cls in onto.classes() if len(cls.instances()) == 0 ] # Busca todas las clases que no tienen instancias.
# No queremos eliminar aquellas que tienen subclases. Vamos a adaptar clases_sin_instancias para que no elimine las clases que tienen subclases.
clases_sin_instancias = [cls for cls in clases_sin_instancias if len(cls.subclasses()) == 0] # Filtra las clases que no tienen subclases.
# Ahora tenemos una lista de clases que no tienen instancias y no tienen subclases. Vamos a eliminarlas de la ontología.

for cls in clases_sin_instancias:
    destroy_entity(cls) # Elimina la clase de la ontología. Esto es útil para reducir el tamaño de la ontología y mejorar su rendimiento.



# for cls in clases_sin_instancias:
#     if len(cls.subclasses()) == 0: # Si la clase no tiene subclases, la eliminamos.
#         destroy_entity(cls) # Elimina la clase de la ontología.


# for cls in clases_sin_instancias:
#     destroy_entity(cls) # Elimina la clase de la ontología.


onto.save(file="SOMA_modificada_podada.owl", format="rdfxml") # Guarda la ontología podada en un nuevo archivo.

onto = get_ontology("SOMA_modificada_podada.owl").load() # Carga la ontología podada.

for cls in onto.classes():
    print(cls) # Imprime todas las clases de la ontología podada. Las clases que no tienen instancias no deberían aparecer aquí.
```
